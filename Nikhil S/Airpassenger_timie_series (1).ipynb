```python
plt.xlabel('Date')
plt.ylabel('Number of passenger')
plt.plot(IndexData)
```

```python
rlmean=IndexData.rolling(window=12).mean()
rlstd=IndexData.rolling(window=12).std()
print(rlmean, rlstd)
```

```python
orginal=plt.plot(IndexData, color='red', label ='Og')
mean=plt.plot(rlmean, color='green', label="Rolling mean")
std=plt.plot(rlstd, color='black', label="Rolling Std")
plt.legend(loc='best')
plt.show()
```

```python
#when we use the log
IndexData_logScale=np.log(IndexData)
plt.plot(IndexData_logScale)
```

```python
Movingavg= IndexData_logScale.rolling(window=12).mean()
movingstd= IndexData_logScale.rolling(window=12).std()
plt.plot(IndexData_logScale)
plt.plot(Movingavg, color='red')
```

```python
#  getting time series as stationary
df_Log_minus_moving_avg= IndexData_logScale-Movingavg
df_Log_minus_moving_avg.head(12)
```

```python
#dropping NaN
df_Log_minus_moving_avg.dropna(inplace=True)
df_Log_minus_moving_avg.head(10)
```

```python
 exp_decay_weighted_avg=IndexData_logScale.ewm(halflife=12, min_periods=0, adjust=True).mean()
 plt.plot(IndexData_logScale)
 plt.plot(exp_decay_weighted_avg)
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(IndexData_logScale)

trend= decomposition.trend
seasonal= decomposition.seasonal
residual= decomposition.resid
```

```python
plt.subplot(411)
plt.plot(IndexData_logScale, label='Og')
plt.legend(loc='best')
plt.subplot(412)

plt.plot(trend, label='trend')
plt.legend(loc='best')
plt.subplot(413)

plt.plot(seasonal, label='seasonal')
plt.legend(loc='best')
plt.subplot(414)

plt.plot(residual, label='Residuals')
plt.legend(loc='best')
plt.tight_layout()
```

```python
from statsmodels.tsa.stattools import adfuller
res=adfuller(IndexData)
print(res[0])
print(res[1])
```

```python
IndexData['first_difference']=IndexData['#Passengers']-IndexData['#Passengers'].shift(1)
res=adfuller(IndexData['first_difference'].dropna())
print(res[0])
print(res[1])
```

```python
IndexData['second_difference']=IndexData['first_difference']-IndexData['first_difference'].shift(1)
res=adfuller(IndexData['second_difference'].dropna())
print(res[0])
print(res[1])
```

```python
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
```

Model training

```python
from statsmodels.tsa.arima.model import ARIMA
new_df=IndexData.drop(['first_difference','second_difference'],axis=1)
```

```python
train_data=new_df['1949-01-01':'1960-01-01']
test_data=new_df['1960-02-01':'1960-12-01']
```

```python
model1=ARIMA(train_data,order=(11,2,12))
model2=ARIMA(train_data,order=(12,2,12))

res1=model1.fit()
res2=model2.fit()



aic1=res1.aic
bic1=res1.bic

aic2=res2.aic
bic2=res2.bic
```

```python
print(aic1)
print(aic2)
if(aic1>aic2):
  print('model 2 has lower aic value,hence a better model')
else:
  print('model 1 has lower aic value,hence a better model')
```

```python
print(bic1)
print(bic2)
if(bic1>bic2):
  print('model 2 has lower bic value,hence a better model')
else:
  print('model 1 has lower bic value,hence a better model')
```

```python
model=ARIMA(train_data,order=(11,2,12))
result=model.fit()
```

```python
forecast =result.predict(start='1960-02-01', end='1960-12-01', dynamic=True)
forecast
```

```python
plt.figure(figsize=(15, 8))
plt.plot(IndexData.index, IndexData['#Passengers'], label='Original Data', color='blue')

plt.plot(forecast.index, forecast, label='Forecast', color='red')

plt.title('ARIMA Forecast')
plt.xlabel('Date')
plt.ylabel('Value')

plt.legend()
plt.show()
```
